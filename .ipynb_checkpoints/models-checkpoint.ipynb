```python
import librosa
import librosa.display
import librosa.feature
import glob
import numpy as np
import matplotlib.pyplot as plt

from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.utils.np_utils import to_categorical
```

```python
def display_mfcc(song):
    y, _ = librosa.load(song)
    mfcc = librosa.feature.mfcc(y)
    
    plt.figure(figsize=(10,4))
    librosa.display.specshow(mfcc, x_axis='time', y_axis='mel')
    plt.colorbar()
    plt.title(song)
    plt.tight_layout()
    plt.show()
```

```python
display_mfcc('dataset/hiphop/hiphop.00040.wav')
```

```python
display_mfcc('dataset/classical/classical.00025.wav')
```

```python
display_mfcc('dataset/pop/pop.00025.wav')
```

```python
def extract_features(f):
    y,_ = librosa.load(f)
    
    mfcc = librosa.feature.mfcc(y)
    mfcc /= np.amax(np.absolute(mfcc)) # we want small values
    
    # data for each song needs to have same size
    truncated_mfcc = np.ndarray.flatten(mfcc)[:25000]
    
    return truncated_mfcc
```

```python
def get_features_n_labels():
    all_features = []
    all_labels = []
    
    genres = ['blues','classical','country','disco','hiphop','jazz','metal','pop','reggae','rock']
    
    for genre in genres:
        file_paths = glob.glob(f'dataset/{genre}/*.wav')
        print(f'processing {len(file_paths)} in {genre}')
        
        for path in file_paths:
            song_features = extract_features(path)
            all_features.append(song_features)
            all_labels.append(genre)
            
    # one-hot encoding: change label names to binary
    unique_genres, genre_idxs = np.unique(all_labels, return_inverse=True)
    genre_idxs = genre_idxs.astype(np.int32, copy=False)
    onehot_labels = to_categorical(genre_idxs, len(unique_genres))
    return np.stack(all_features), onehot_labels
```

```python
features, labels = get_features_n_labels()
```

```python
# split dataset for training and test
print(f'features: {features.shape}')
print(f'labels: {labels.shape}')
```

```python
# combine features and labels into single matrix
dataset = np.column_stack((features, labels))

# shuffle data
np.random.shuffle(dataset)

# split data
train_split = 0.8
split_idx = int(len(dataset) * train_split)
train, test = dataset[:split_idx, :], dataset[split_idx:, :]

print(f'train: {train.shape}')
print(f'test: {test.shape}')

# split features and labels 
train_features = train[:,:-10]
train_labels = train[:,-10:]
test_features = test[:,:-10]
test_labels = test[:,-10:]

print(f'train_features: {train_features.shape}')
print(f'train_labels: {train_labels.shape}')
```

```python
# build model
# TODO: need to play around with layers
model = Sequential([
    Dense(100, input_dim=train_features.shape[1]),
    Activation('relu'),
    Dense(10),
    Activation('softmax'),
])

# TODO: reasons for specific optimizer, loss function
model.compile(optimizer='adam',
             loss='categorical_crossentropy',
             metrics=['accuracy'])
print(model.summary())

model.fit(train_features, train_labels, epochs=10, batch_size=32,
         validation_split=0.2)

loss, acc = model.evaluate(test_features, test_labels, batch_size=32)
print(f'loss: {loss}, accuracy: {acc}')
```
